# Random Forest Project 

For this project we will be exploring publicly available data from [LendingClub.com](www.lendingclub.com). Lending Club connects people who need money (borrowers) with people who have money (investors). Hopefully, as an investor you would want to invest in people who showed a profile of having a high probability of paying you back. We will try to create a model that will help predict this.

We will use lending data from 2007-2010 and be trying to classify and predict whether or not the borrower paid back their loan in full. You can download the data from [here](https://www.lendingclub.com/info/download-data.action) or just use the csv already provided. It's recommended you use the csv provided as it has been cleaned of NA values.

Here are what the columns represent:
* credit.policy: 1 if the customer meets the credit underwriting criteria of LendingClub.com, and 0 otherwise.
* purpose: The purpose of the loan (takes values "credit_card", "debt_consolidation", "educational", "major_purchase", "small_business", and "all_other").
* int.rate: The interest rate of the loan, as a proportion (a rate of 11% would be stored as 0.11). Borrowers judged by LendingClub.com to be more risky are assigned higher interest rates.
* installment: The monthly installments owed by the borrower if the loan is funded.
* log.annual.inc: The natural log of the self-reported annual income of the borrower.
* dti: The debt-to-income ratio of the borrower (amount of debt divided by annual income).
* fico: The FICO credit score of the borrower. ( Lenders use borrowers' FICO scores along with other details on borrowers' credit reports to assess credit risk and determine whether to extend credit.)
* days.with.cr.line: The number of days the borrower has had a credit line.
* revol.bal: The borrower's revolving balance (amount unpaid at the end of the credit card billing cycle).
* revol.util: The borrower's revolving line utilization rate (the amount of the credit line used relative to total credit available).
* inq.last.6mths: The borrower's number of inquiries by creditors in the last 6 months.
* delinq.2yrs: The number of times the borrower had been 30+ days past due on a payment in the past 2 years.
* pub.rec: The borrower's number of derogatory public records (bankruptcy filings, tax liens, or judgments).
* not.fully.paid: the output label column. not.fully.paid=0 means customer is paying to the fully

## Categorical Features

Notice that the **purpose** column is categorical. That means we need to transform them using dummy variables so the model will be able to understand them. Let's do this in one clean step using pd.get_dummies.

Let's deal with these columns in a way that can be expanded to multiple categorical features if necessary.

## Training a Decision Tree Model

Let's start by training a single decision tree first!

**Import DecisionTreeClassifier**

```python
dtree = DecisionTreeClassifier(random_state=101)
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=101
)
dtree.fit(x_train, y_train)
```

### 2- Let's train now another RFmodel by tuning its hyperparameters!
 
1- Create an instance of RandomForestClassifier(random_state=2026) called **RFGrid**
 
2- Create a dictionary named **param_grid** containig the following keys and values:
 
     - 'class_weight': [{0:1, 1:10},{0:1, 1:5}],
     - 'max_depth': [20, 10, 15],
     - 'min_samples_split': [2, 5, 20],
     - 'min_samples_leaf': [2, 20],
     - 'ccp_alpha': [0.003, 0.002],
     - 'n_estimators':[100,150]
   Research and learn the role of each parameter
3- Create a dictionary called **scoring** containing the F1_score of the minority class as a main scoring metric and the balanced_accuracy_score as second. 
- Use the **make_scorer** function from sklearn to order your scoring.
 
4- Create an instance called **grid** of the GridsearchCV model from sklearn. 
- Use 5-fold CV and the 2 dictionaries you have just created. 
- Select the best parameters based on the F1_score
 
5- Fit this model on your training set
 
6- Print the best parameters obtained by your search
 
7- Print the F1_score and the balanced_accuracy_score obtained

```python
#part 1
RFGrid = RandomForestClassifier(random_state=2026)
```

```python
#part 2
from sklearn.model_selection import GridSearchCV
param_grid = {
 'class_weight': [{0:1, 1:10}, {0:1, 1:5}], #used for importance of classes(the minority class is 10 times more important than the majority class in the first case and 5 times in the second case)
 'max_depth': [20, 10, 15], #max depth of trees (additional nodes will be added until the max depth is reached)
 'min_samples_split': [2, 5, 20], #minimum number of samples required to split a node (pre-pruning)
 'min_samples_leaf': [2, 20], #minimum number of samples required to be at a leaf node(pre-pruning)
 'ccp_alpha': [0.003, 0.002], #Cost Complexity Pruning parameter. Helps reduce overfitting. Removes leaves that have a cost complexity greater than the given value.(post-pruning)
 'n_estimators': [100,150] #number of trees in the forest
}
```

```python
from sklearn.metrics import make_scorer, f1_score, balanced_accuracy_score

#part 3
scoring = {
    'f1_minority': make_scorer(f1_score, pos_label=1), #pos_label=1 because we want to focus on the minority class
    'balanced_accuracy': make_scorer(balanced_accuracy_score)
}
#Balance accuracy is used because of the imbalance in the dataset. It gives a better measure of performance when classes are imbalanced.
```

```python
#part 4
grid = GridSearchCV(
    estimator=RFGrid, #the model we want to tune
    param_grid=param_grid, #the parameters we want to try
    scoring=scoring,
    refit='f1_minority', #the metric to use for selecting the best model
    cv=5 #the number of cross-validation folds (run 5 times with different splits of the data)
)
```

```python
#part 5
grid.fit(x_train, y_train)
```

```python
#part 6
grid.best_params_
#used for the best model that we chose in refit='f1_minority'
#so f1 here
```

```python
#part 7
print("Best F1_score:", grid.cv_results_['mean_test_f1_minority'][grid.best_index_])
print("Best Balanced Accuracy:", grid.cv_results_['mean_test_balanced_accuracy'][grid.best_index_])
#we can use to print also
print("Best F1-score:",grid.best_score_)
y_pred_best=grid.best_estimator_.predict(x_test)
print("Balanced accuracy:",balanced_accuracy_score(y_test,y_pred_best))
#the answees are scored in grid.cv_results_ and we access the mean_test_f1_minority and mean_test_balanced_accuracy for the best index which is the best model according to f1_minority.
```
