## Deep Learning Model Selection

Three deep learning candidates are considered for this binary classification task: a **shallow neural network**, a **medium neural network**, and a **deeper neural network**.

Since the dataset contains only **6 structured features**, the goal is to test simple fully connected networks first. A convolutional neural network is not selected here because CNNs are mainly useful when the input has spatial structure, such as images, while this dataset is tabular.


### Model 1: Shallow Neural Network


```python
import tensorflow as tf

def weight(shape, stddev=0.1):
    return tf.Variable(tf.random.normal(shape, stddev=stddev))

def bias(shape):
    return tf.Variable(tf.zeros(shape))

n_features = 6
W1_shallow = weight((n_features, 8))
b1_shallow = bias((8,))
W2_shallow = weight((8, 1))
b2_shallow = bias((1,))

vars_shallow = [W1_shallow, b1_shallow, W2_shallow, b2_shallow]

def forward_shallow(x):
    h = tf.nn.relu(tf.matmul(x, W1_shallow) + b1_shallow)
    logits = tf.matmul(h, W2_shallow) + b2_shallow
    return logits

print("Shallow Neural Network created successfully")
print(f"Hidden layers: 1")
print(f"Hidden units: 8")
print(f"Input features: {n_features}")
```

### Model 2: Medium Neural Network


```python
W1_medium = weight((n_features, 16))
b1_medium = bias((16,))
W2_medium = weight((16, 8))
b2_medium = bias((8,))
W3_medium = weight((8, 1))
b3_medium = bias((1,))

vars_medium = [W1_medium, b1_medium, W2_medium, b2_medium, W3_medium, b3_medium]

def forward_medium(x):
    h1 = tf.nn.relu(tf.matmul(x, W1_medium) + b1_medium)
    h2 = tf.nn.relu(tf.matmul(h1, W2_medium) + b2_medium)
    logits = tf.matmul(h2, W3_medium) + b3_medium
    return logits

print(f"Input features: {n_features}")
print(f"Hidden layers: 2")
print(f"Hidden units: 16 -> 8")
print(f"Input features: {n_features}")
```

### Model 3: Deeper Neural Network


```python
W1_deep = weight((n_features, 32))
b1_deep = bias((32,))
W2_deep = weight((32, 16))
b2_deep = bias((16,))
W3_deep = weight((16, 8))
b3_deep = bias((8,))
W4_deep = weight((8, 1))
b4_deep = bias((1,))

vars_deep = [W1_deep, b1_deep, W2_deep, b2_deep, W3_deep, b3_deep, W4_deep, b4_deep]

def forward_deep(x, training=True):
    h1 = tf.nn.relu(tf.matmul(x, W1_deep) + b1_deep)
    if training:
        h1 = tf.nn.dropout(h1, rate=0.2)
    h2 = tf.nn.relu(tf.matmul(h1, W2_deep) + b2_deep)
    h3 = tf.nn.relu(tf.matmul(h2, W3_deep) + b3_deep)
    logits = tf.matmul(h3, W4_deep) + b4_deep
    return logits

print("Deeper Neural Network created successfully")
print(f"Hidden layers: 3")
print(f"Hidden units: 32 -> 16 -> 8")
print(f"Dropout: 0.2")
print(f"Input features: {n_features}")
```

## Selected Models

**The medium neural network** is the most reasonable starting choice because it adds some non-linearity without becoming too large for a dataset with only 6 input features.

**The shallow neural network** is kept as a baseline because it is the simplest deep learning option and helps show whether extra depth actually improves performance.

**The deeper neural network** is included as a comparison model, but it may be more complex than necessary for this problem and could overfit more easily.

A **CNN is not selected** because the input is tabular rather than image-based. If the project later uses resume images or page layouts directly, then a CNN would make more sense.
